```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Load the dataset
data = pd.read_csv('Position_Salaries.csv')

# Separate the independent variable (Level) and the dependent variable (Salary)
X = data[['Level']].values
y = data['Salary'].values

# Transform features to polynomial form (degree 2 or higher as needed)
poly = PolynomialFeatures(degree=2)  # You can increase degree for more complexity
X_poly = poly.fit_transform(X)

# Fit a linear regression model on the transformed polynomial features
model = LinearRegression()
model.fit(X_poly, y)

# Predict using the model
y_pred = model.predict(X_poly)

# Plot the results
plt.scatter(X, y, color='blue', label='Actual Salary')
plt.plot(X, y_pred, color='red', label='Polynomial Regression Prediction')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Polynomial Regression for Position Salaries')
plt.legend()
plt.show()

# Print model parameters
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```
